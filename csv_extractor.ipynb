This script takes all csv files from Batch 1 (historical data) and put it into dataframe.

```python
import os
from pathlib import Path
import glob
from pyspark.sql import SparkSession
import time

spark = SparkSession.builder.appName("TPC-DI Historical Data").getOrCreate()
```
